# Assignment 17 - Letters Count
## Objective
This program takes a sentence from the user and counts the number of each character in the sentence.

## Method
- Take input from the user
- Loop through each character
- Store the count in a dictionary

```python
sentence = input("Enter a sentence: ")

char_count = {}

for char in sentence:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```
